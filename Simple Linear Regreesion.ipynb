# Salary_hike -> Build a prediction model for Salary_hike ------------------------------------------------------------ Build a simple linear regression model by performing EDA and do necessary transformations and select the best model using R or Python.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from numpy.polynomial.polynomial import polyfit
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("C:/Users/Admin/Downloads/Salary_Data.csv")
df
```

# Scatter Plot

```python
x = df['YearsExperience']
y = df['Salary']
```

```python
b, m = polyfit(x, y, 1)
plt.scatter(x, y)
plt.plot(x, y, '.')
plt.plot(x, b + m * x, '-')
plt.title('Scatter plot Salary Hike')
plt.xlabel('Years of Experience')
plt.ylabel('Salary Hike')
plt.show()
```

# Correlation Analysis

```python
corr = np.corrcoef(x, y)
```

# Regression Model

```python
model = sm.OLS(y, x).fit()
predictions = model.predict(x)
model.summary()
```

```python
x_log = np.log(df['YearsExperience'])
```

```python
model = sm.OLS(y, x_log).fit()
predictions = model.predict(x_log)
model.summary()
```

```python
y_log = np.log(df['Salary'])
```

```python
model = sm.OLS(y_log, x).fit()
predictions = model.predict(x)
model.summary()
```

```python
model = sm.OLS(y_log, x_log).fit()
predictions = model.predict(x_log)
model.summary()
```

```python
x_sqrt = np.sqrt(df['YearsExperience'])
```

```python
model = sm.OLS(y, x_sqrt).fit()
predictions = model.predict(x_sqrt)
model.summary()
```

```python
y_sqrt = np.sqrt(df['Salary'])
```

```python
model = sm.OLS(y_sqrt, x).fit()
predictions = model.predict(x)
model.summary()
```

```python
model = sm.OLS(y_sqrt, x_sqrt).fit()
predictions = model.predict(x_sqrt)
model.summary()
```
